## Model

```python
enc_out_size=300
dec_out_size=64 
num_dec_layers=3
```

```python
%xdel generator
%xdel decoder
%xdel encoder
encoder = Encoder(len(vocab), enc_out_size)
decoder = Decoder(dec_out_size, enc_out_size, num_dec_layers)
generator = MGenerator(encoder, decoder)
```

## Learner and Train

```python
%xdel learn
learn = get_generator_learner(dls, generator)
```

```python
learn.fit_one_cycle(1, 0.069)
```
